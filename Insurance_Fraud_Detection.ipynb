# 1 - Support Vector Classifer

**ðŸ”¹ What is a Support Vector Classifier (SVC)?**
A Support Vector Classifier (SVC) is a supervised learning algorithm used for binary and multi-class classification. It is based on Support Vector Machines (SVM) and aims to find the optimal decision boundary that maximizes the margin between different classes.

**1ï¸âƒ£ Intuition Behind SVC**
SVC tries to find a hyperplane that separates different classes with the maximum margin. The margin is the distance between the hyperplane and the closest data points (called support vectors).

For a 2D dataset â†’ The decision boundary is a line.

For a 3D dataset â†’ The decision boundary is a plane.

For an n-dimensional dataset â†’ The decision boundary is a hyperplane.

If the data is not linearly separable, SVC uses the kernel trick to project the data into a higher dimension where it becomes separable.



4ï¸âƒ£ Hyperparameters in SVC
ðŸ”¹ Key Hyperparameters:
C (Regularization Parameter)

Controls the trade-off between maximizing the margin and allowing some misclassification.

High C â†’ Less margin, fewer misclassifications (Overfitting).

Low C â†’ Larger margin, more misclassifications (Underfitting).

kernel (Kernel Function)

linear â†’ Simple and works well with linearly separable data.

poly â†’ For polynomial decision boundaries.

rbf (default) â†’ For non-linearly separable data.

sigmoid â†’ Similar to neural networks.

gamma (for RBF and Polynomial Kernels)

Defines how much influence a single training example has.

High gamma â†’ Model becomes more complex (Overfitting).

Low gamma â†’ Model becomes simple (Underfitting).

degree (for Polynomial Kernel)

Defines the degree of the polynomial kernel.

Only applicable when kernel='poly'.

probability (Boolean)

If True, enables probability estimates using cross-validation.

max_iter

Maximum number of iterations for training.

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

svc_train_acc=accuracy_score(y_train, svc.predict(x_train))
svc_test_acc=accuracy_score(y_test, y_pred)

print(f"Training Accuracy: {svc_train_acc}")
print(f"Testing Accuracy: {svc_test_acc}")


print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train, y_train)

y_pred=dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dtc_train_acc=accuracy_score(y_train, dtc.predict(x_train))
dtc_test_acc=accuracy_score(y_test, y_pred)

print(f"Training Accuracy: {dtc_train_acc}")
print(f"Testing Accuracy: {dtc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# hyper parameter tuning

from sklearn.model_selection import GridSearchCV

grid_param={
    'criterion':['gini', 'entropy'],
    'max_depth':[3,5,7,10],
    'min_samples_leaf':range(1,10,1),
    'min_samples_split':range(2,10,1)
}

grid_search=GridSearchCV(estimator=dtc, param_grid=grid_param, cv=5, n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
# best parameters and best score

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
# best estimator

dtc = grid_search.best_estimator_

y_pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

dtc_train_acc=accuracy_score(y_train, dtc.predict(x_train))
dtc_test_acc=accuracy_score(y_test, y_pred)

print(f"Training Accuracy: {dtc_train_acc}")
print(f"Testing Accuracy: {dtc_test_acc}")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
